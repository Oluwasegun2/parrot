# Cleaning the column headers of all tables using a snake-case function

```python
def to_snake_case(name: str) -> str:
    # Replace camelCase or PascalCase with snake_case
    name = re.sub(r'(?<!^)(?=[A-Z])', '_', name)
    # Replace spaces, hyphens, and dots with underscores, then lowercase
    name = re.sub(r'[\s\-.]+', '_', name).lower()
    return name.strip('_')

def clean_column_names(df: DataFrame) -> DataFrame:
    new_columns = {col: to_snake_case(col) for col in df.columns}
    df_cleaned = df.select([df[col].alias(new_columns[col]) for col in df.columns])
    return df_cleaned, list(new_columns.values())
```

```python
table_name = []

for i in dbutils.fs.ls('mnt/silver/'):
    table_name.append(i.name.split('/')[0])

table_name
```

# Saving the cleaned data into the Gold layer 

```python
for name in table_name:
    path = '/mnt/silver/' + name
    print(path)
    df = spark.read.format('parquet').load(path)

    df = clean_column_names(df)[0]

    output_path = '/mnt/gold/' + name + '/'
    df.write.format('delta').mode('overwrite').save(output_path)
```
